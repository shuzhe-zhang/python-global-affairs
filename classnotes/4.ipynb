```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up the WebDriver (e.g., for Chrome)
driver = webdriver.Chrome()

# Open the web page
driver.get("https://www.aeaweb.org/journals/aer/issues")  # Replace 'YourPageURL' with the actual URL

# Find all issue link elements on the page
issue_links = driver.find_elements(By.CSS_SELECTOR, "a[href*='issues']")  # Adjust the CSS selector as needed based on actual HTML
print(issue_links)
# Loop through each link and click on it
for link in issue_links:
    link.click()
    time.sleep(2)  # Wait for 2 seconds or until the page loads completely
    # You may want to perform additional actions on each issue page here
    
    # Navigate back to the main page to find the issues again
    driver.back()

# Close the browser
driver.quit()
```

```python
def get_data(url) -> list:
    browser_options = ChromeOptions()
    driver = Chrome(options = browser_options)
    driver.get(url) 
    
    element = driver.find_element(By.LINK_TEXT, "Science Fiction")
    element.click()
    
    
    books = driver.find_elements(By.CSS_SELECTOR, ".product_pod")
    
    data = []
    for book in books:
        title = book.find_element(By.CSS_SELECTOR, "h3 > a")
        price = book.find_element(By.CSS_SELECTOR, ".price_color")
        stock = book.find_element(By.CSS_SELECTOR, ".instock.availability")
        
        book_items = {
            "title": title.get_attribute("title"),
            "price": price.text,
            "stock": stock.text
        }
        data.append(book_items)
    driver.quit()
    return data

def main():
    data = get_data("https://books.toscrape.com/")
    print(data)
    
main()
```
